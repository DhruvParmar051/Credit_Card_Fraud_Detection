üöÄ Understanding the Problem

Credit card fraud is a growing concern in the financial industry.  
This project aims to detect fraudulent transactions using machine learning techniques.  
We will analyze various transaction features and train models to classify transactions as 'Fraudulent' or 'Legitimate'.

----


üìä Data Collection
- Dataset Source: [Kaggle - Credit Card Fraud Detection](https://www.kaggle.com/mlg-ulb/creditcardfraud)  
- The dataset consists of 284,807 transactions with 30 features, including:  
  - `Time`: Time elapsed since the first transaction.  
  - `V1` to `V28`: Features transformed using PCA.  
  - `Amount`: Transaction amount.  
  - `Class`: Target variable (0 = Legitimate, 1 = Fraudulent).

1)  Data Collection


```python
# import Required Packages

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
csv_path = "data\creditcard.csv"
credit_df = pd.read_csv(csv_path)
print(credit_df.head())
print(credit_df.shape)
```

2. Data Checks to Perform

```python
# list the columns of credit_df
print(credit_df.columns)
```

```python
# Check for missing values
credit_df.isna().sum()
```

```python
# check for duplicate
credit_df.duplicated().sum()
```

```python
# remove duplicates
credit_df = credit_df.drop_duplicates()
```

```python
# check for data type
credit_df.info()
```

```python
# check for unique values
credit_df.nunique()
```

```python
# Check for statistical description
credit_df.describe()
```

4. Exploring Data ( Visualization )

```python
# Fraud vs Non-Fraud Transactions
sns.countplot(data=credit_df, x='Class')
plt.title("Fraud vs Non-Fraud Transactions")
plt.xticks([0, 1], ['Non-Fraud', 'Fraud'])
plt.show()
```

```python
# Transaction Amount Distribution (Fraud vs Non-Fraud)
sns.histplot(data=credit_df, x='Amount', hue='Class', bins=50, kde=True)
plt.xscale('log')  # Log scale helps if Amount has a large range
plt.title("Transaction Amount Distribution (Fraud vs Non-Fraud)")
plt.show()
```

```python
# Transaction Amounts by Class
sns.boxplot(data=credit_df, x='Class', y='Amount')
plt.yscale('log')  # Helps visualize outliers better
plt.title("Transaction Amounts by Class")
plt.show()
```

```python
# Feature Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(credit_df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Transaction Amount Over Time (Fraud vs Non-Fraud)
sns.scatterplot(data=credit_df, x='Time', y='Amount', hue='Class', alpha=0.5)
plt.xscale('log')
plt.title("Transaction Amount Over Time (Fraud vs Non-Fraud)")
plt.show()
```

```python
# Pie Chart: Fraud vs Non-Fraud Transactions
# Count occurrences of each class
fraud_counts = credit_df['Class'].value_counts()

# Labels and colors
labels = ['Non-Fraud', 'Fraud']
colors = ['lightblue', 'red']

# Plot Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(fraud_counts, labels=labels, autopct='%1.2f%%', colors=colors, startangle=45, explode=[0, 0.1])
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```
